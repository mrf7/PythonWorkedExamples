```python
import json
from pprint import pprint 
# Open the data file 
file = open("cs1064_worked_examples_data.json")
data = json.load(file)
```

## Break down of number of students that attempted and completed the problems

```python
import numpy as np
import matplotlib.pyplot as plt
labels = "Attempted", "Completed", "Not Attempted"
# Holds data for each assignment 
f17Data =[
  {'name': '#43.4) Multiple Returns',
    'attempted': 0,
    'completed': 0,
    'not_attempted': 0},
 {'name': '#40.2) Word Frequency',
  'attempted': 0,
  'completed': 0,
  'not_attempted': 0},
 {'name': '#39.4) Default Name',
  'attempted': 0,
  'completed': 0,
  'not_attempted': 0},
 {'name': '#33.3) Cube Elements',
  'attempted': 0,
  'completed': 0,
  'not_attempted': 0}]

s18Data =[
  {'name': '#43.4) Multiple Returns',
    'attempted': 0,
    'completed': 0,
    'not_attempted': 0},
 {'name': '#40.2) Word Frequency',
  'attempted': 0,
  'completed': 0,
  'not_attempted': 0},
 {'name': '#39.4) Default Name',
  'attempted': 0,
  'completed': 0,
  'not_attempted': 0},
 {'name': '#33.3) Cube Elements',
  'attempted': 0,
  'completed': 0,
  'not_attempted': 0}]

for student in data:
    assignments = student["assignment"]
    problems = f17Data if student["user"]["section"] == "Fall 2017" else s18Data
    for problem in problems:
        assign = assignments[problem["name"]]
        if assign["correct"]:
            problem["completed"] = problem["completed"]+ 1
        elif assign["attempted"]:
            problem["attempted"] = problem["attempted"] + 1
        else:
            problem["not_attempted"] = problem["not_attempted"] + 1
            
labels = "attempted", "completed", "not_attempted"
for i in range(0,4):
    problem = f17Data[i]
    fig1, ax1 = plt.subplots()
    sizes = [problem["attempted"], problem["completed"], problem["not_attempted"]]
    ax1.pie(sizes, labels=labels, autopct="%.2f")
    ax1.set_title("F17 " + problem["name"])
    ax1.axis('equal')
    problem = s18Data[i]
    fig1, ax1 = plt.subplots()
    sizes = [problem["attempted"], problem["completed"], problem["not_attempted"]]
    ax1.pie(sizes, labels=labels, autopct="%.2f")
    ax1.set_title("S18 " + problem["name"])
    ax1.axis('equal')
plt.show()
```
